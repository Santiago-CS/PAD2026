## CREACIÓN DE BASES DE DATOS SINTÉTICAS

### La Librería Faker
<center><img src="./Imagenes/Faker.png" style="width:500px;height:117px;" class="center"></center>

**Faker** es un *paquete* de **Python** del tipo ***open source*** para la generación de datos falsos que pueden ser agregados a un **DataFrame** de **Pandas**, es de gran utilidad para hacer pruebas de modelos así como para anonimizar información sensible.

Para instalar la librería se emplea la siguiente instrucción en una ***Terminal (Shell)***:

```python
pip install Faker
```

Una vez instalada, la librería se importará por medio de la función:
```python
from faker import Faker
```
A partir de ello se podrá hacer uso de los distintos métodos de **Faker** invocándolos por medio de `faker`.

La información y documentación de **Faker** se puede encontrar en su portal web a través [de esta liga](https://pypi.org/project/Faker/0.7.4) y su documentación está disponible a través [de esta liga](https://faker.readthedocs.io/en/master).

**Faker** hace uso de ***Providers***, esto es, información específica para locaciones mundiales donde cada uno de los lugares puede generar datos falsos para ciertas etiquetas en particular. Los ***Providers*** disponibles pueden ser consultados [en esta liga](
https://faker.readthedocs.io/en/master/locales.html).

#### Implementación de un Ejemplo
Se generará un **DataFrame** con datos falsos pero que sean relacionados con **México**, para ello se empleará el ***Provider*** con datos locales de **México** a través de la etiqueta `es_MX`, con la suposición de que será una lista de clientes que han realizado la compra de artículos en una tienda en línea.

Primeramente se realizarán las siguientes operaciones:

1. Importación de las librerías necesarias (**Pandas**, **DateTime**, **Random** y **Faker**).
2. Se define una variable con la cantidad de datos falsos a ser generados (`no_clientes`).
3. Se define el ***Provider*** con datos locales de **México** (a través de la etiqueta `es_MX`).
4. Se inicializa una lista donde se agregarán los elementos falsos que se hayan generado, esta lista se pasará a un **DataFrame**.

````python
#Importación de Librerías
import random
import pandas as pd
from faker import Faker
from datetime import datetime

#Definición de la Cantidad de Datos a Generar
no_clientes = 1000

#Definición del Tipo de Provider a Emplear
fake = Faker(locale = 'es_MX')

#Inicialización de la Lista para los Datos Falsos
clientes = []
````

Los distintos tipos de datos falsos que pueden ser generados para **México** se pueden consultar [en esta liga](https://faker.readthedocs.io/en/master/locales.html).

Ahora se realizará un ciclo `for` a través del cual de generarán los datos falsos que corresponden a:
1. Fecha de la transacción por medio del método `.date_between()`.
2. Nombre de la persona por medio del método `.name()`.
3. Género de la persona por medio del método `.choice()` de la librería **Random**.
4. Clave Única de Registro de Población (***CURP***) de la persona por medio del método `.curp()`.
5. Clave del Elector de la Identificación Oficial de la persona por medio del método `.elector_code()`.
6. Registro Federal del Contribuyente (***RFC***) de la persona por medio del método `.rfc()`.
7. Número de Seguro Social (***NSS***) de la persona por medio del método `.ssn()`.
8. Número de Teléfono de la persona por medio del método `.phone_number()`.
9. Correo Electrónico de la persona por medio del método `.ascii_email()`.
10. Ciudad donde vive la persona por medio del método `.city()`.
11. Estado donde vive la persona por medio del médoto `.state()`.
12. Código Postal del lugar donde vive la persona por medio del método `.postcode()`.
13. Número de Identificación del producto comprado (código de barras de 13 dígitos) por medio del método `.ean()`.
14. Cantidad Pagada por el producto comprado por medio del método `.pyfloat()`.
15. Banco a través del cual se realizó el pago por medio del método `.bank()`.

***Nota Importante:*** Los valores son generados de manera aleatoria por medio de la librería **Faker**, sin embargo no existe una relación entre ellos. Por ejemplo, el ***CURP*** y el ***RFC*** no estarán relacionados directamente con el nombre de la persona.

````python
#Ciclo FOR para la Generación de los Valores Falsos
for variable in range(no_clientes):

    #Crear la Fecha de la Compra 
    d1 = datetime.strptime(f'1/1/2024', '%d/%m/%Y')
    d2 = datetime.strptime(f'31/12/2024', '%d/%m/%Y')
    fecha = fake.date_between(d1, d2)

    #Crear el Nombre del Cliente
    nombre = fake.name()

    #Crear el Género del Cliente
    genero = random.choice(["H", "M"])

    #Crear el CURP del Cliente
    curp = fake.curp()

    #Crear la Clave del Elector
    ce = fake.elector_code()

    #Crear el RFC del Cliente
    rfc = fake.rfc()

    #Crear el NSS del Cliente
    nss = fake.ssn()
    
    #Crear el Número de Teléfono
    telefono = fake.phone_number()
    
    #Crear el Correo Electrónico
    correo = fake.ascii_email()

    #Crear la Ciudad
    ciudad = fake.city()

    #Crear el Estado
    estado = fake.state()

    #Crear el Código Postal
    cp = fake.postcode()

    #Crear el ID del Producto 
    id_producto = fake.ean(length = 13)
    
    #Crear el Costo del Producto
    costo = fake.pyfloat(right_digits = 2, positive = True, min_value = 1, max_value = 5000)

    #Crear el Banco del Pago
    banco = fake.bank()

    #Agregar los Datos a la Lista
    clientes.append([fecha, 
                     nombre, 
                     genero, 
                     curp, 
                     ce,
                     rfc, 
                     nss, 
                     telefono, 
                     correo, 
                     ciudad, 
                     estado,
                     cp,
                     id_producto,
                     costo,
                     banco])
````

Ahora la lista `clientes` que fue generada con datos falsos se convertirá en un **DataFrame** de **Pandas** por medio del método `.DataFrame()`, adicionalmente se mostrarán los resultados y se guardarán en un archivo `CSV`.

````python
#Generación del DataFrame a Partir de la Lista
clientes_df = pd.DataFrame(clientes, columns=['Fecha',
                                              'Nombre', 
                                              'Género',
                                              'CURP',
                                              'Clave del Elector',
                                              'RFC',
                                              'NSS',
                                              'Teléfono',
                                              'Correo Electrónico', 
                                              'Ciudad',
                                              'Estado',
                                              'Código Postal',
                                              'ID Producto', 
                                              'Costo',
                                              'Banco']) 

#Impresión de los Resultados
print("El DataFrame con Datos Falsos es:")
display(clientes_df)

#Guardado de los Datos Falsos a un archivo CSV
clientes_df.to_csv('./Guardados/Falsos_DF.csv')
````

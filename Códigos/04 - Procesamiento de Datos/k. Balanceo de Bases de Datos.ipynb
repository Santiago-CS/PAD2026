## BALANCEO DE BASES DE DATOS

### La Librería Imbalanced-Learn
<center><img src="./Imagenes/Imbalanced_Learn.png" style="width:500px;height:191px;" class="center"></center>

**Imbalanced-Learn** es una *librería* de **Python** del tipo ***open source*** y con licencia del Instituto de Tecnología de Massachusetts (Massachusetts Institute of Technology, MIT) basada en **Scikit-Learn** la cual proporciona herramientas para realizar un balance de los datos contenidos en un **DataFrame** o en una base de datos con fines de clasificación empleando modelos predictivos.

El reto al trabajar con una base de datos no balanceada es que la mayoría de los modelos de **Aprendizaje Automático** tendrán un desempeño muy bajo en las clases minoritarias, y los resultados tendrán un sesgo hacia las clases mayoritarias.

Algunas alternativas para realizar el balanceo se enfocan en un sobre-muestreo de las clases minoritarias, puede ser a través de la duplicación de datos aunque esto no incluye datos nuevos al modelo. Por lo mismo, es posible generar nuevas muestras sintetizadas de las existentes en la base de datos, lo cual es una forma de **Aumentación de Datos** ([***Data Augmentation***](https://machinelearningmastery.com/how-to-configure-image-data-augmentation-when-training-deep-learning-neural-networks)) para las clases minoritarias.

Para instalar la librería **Imbalanced-Learn** se emplea la siguiente instrucción en una ***Terminal (Shell)***:

```python
pip install imbalanced-learn
```

Una vez instalada, la librería se importará por medio de la función:
```python
from imblearn.over_sampling import method
```

La información y documentación de **Imbalanced-Learn** se puede encontrar en su portal web a través [de esta liga](https://imbalanced-learn.org).

#### Implementación de un Ejemplo

Para ejemplificar el **Balanceo** de bases de datos desbalanceadas, se empleará una base de datos que contiene 20,000 registros de dos series y la clase a la que pertenece cada par ordenado de valores. Los encabezados de las columnas de información son:

1. **Serie1**: Es la primera serie de datos [*flotante*].
2. **Serie2**: Es la segunda serie de datos [*flotante*].
3. **Clase**:  Es la ***clase*** para cada par ordenado, se etiquetan como `0` y `1`.

Para la lectura del archivo y la preparación de los datos, se realizarán las siguientes operaciones:

1. Importación de las librerías necesarias (**Pandas**, **NumPy**, **Seaborn**, **Matplitlib** y **Collections**).
2. Definición de los parámetros a emplear en los gráficos de **Matplotlib**.
3. Lectura de los datos empleando el método `.read_csv()`.
4. Mostrar el **DataFrame** resultante empleando la función `display()`.

````python
#Importación de Librerías
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter

#Definición de los parámetros de los gráficos
plt.rcParams.update({'font.size': 10, 'figure.figsize': (8, 6)}) 

#Lectura de los datos y asignación del nombre de las películas como el índice
nobal_df = pd.read_csv("./Datos/No_Balanceados.csv")

#Impresión de los Resultados
print("El DataFrame es:")
display(nobal_df)
````

Estos valores pueden ser revisados de manera sencilla en Python. Sin embargo, una manera rápida es comprobarlo a través de un gráfico, para ello se empleará el método `.scatterplot()` de **Seaborn** aplicado al **DataFrame**, esto es:

````python
#Gráfico de los Valores del DataFrame
sns.scatterplot(x = 'Serie1', y = 'Serie2', hue = 'Clase', data = nobal_df)
plt.show();
````

Ahora se revisará la cantidad de datos que corresponden a cada una de las clases etiquetadas como `0` y `1` en el **DataFrame** por medio del método `.value_counts()` de **Pandas**.

````python
#Conteo de las Clases en el DataFrame
nobal_df['Clase'].value_counts()
````

Finalmente, para preparar los datos para el **Balanceo**, se crearán los siguientes arreglos:
- Un **DataFrame** `x` que contendrá a las dos series de datos.
- Una **Serie** `y` que contendrá las clases se las series de datos.

````python
#Preparación de los Arreglos
x = nobal_df.drop('Clase', axis=1)
y = nobal_df['Clase']
````

### Técnica `SMOTE` para el Balanceo de Datos
La técnica `SMOTE` (Ténica Sintética de Sobre-Muestreo de Minorías, ***Synthetic Minority Oversampling Technique***) es posiblemente la técnica más empleada para sintetizar nuevas muestras, fue desarrollada por Nitesh Chawla en 2002 y descrita en una publicación ue puede ser consultada [en esta liga](https://arxiv.org/abs/1106.1813).

Para su implementación, se realizarán los siguientes pasos:
1. Importación del modelo `SMOTE` de la librería **Imbalanced-Learn**.
2. Se muestra la cantidad de datos para cada clase por medio de la función `Counter()` de la librería **Collections**.
3. Se realiza el sobre-muestreo de los datos por medio del modelo `SMOTE` así como de la función `.fit_resample()`.
4. Se muestra la cantidad de datos para cada clase por medio de la función `Counter()` de la librería **Collections**.
5. Se genera un **DataFrame** balanceado con los nuevos datos generados.

````python
#Importación del Modelo
from imblearn.over_sampling import SMOTE

#Impresión de la Cantidad de Datos por Clase Antes del Balanceo
print('Antes:', Counter(y))

#Sobre-Muestreo de los Datos Desbalanceados
smote = SMOTE()
X_train, y_train = smote.fit_resample(x, y)

#Impresión de la Cantidad de Datos por Clase Después del Balanceo
print('Después:',Counter(y_train))

#Nuevo DataFrame Balanceado
smote_df = X_train.copy()
smote_df['Clase'] = y_train
````

### Técnica `ADASYN` para el Balanceo de Datos
La técnica `ADASYN` (Sintéticos Adaptativos, ***Adaptive Synthetic***) es una variante avanzada de `SMOTE`, sin embargo es probable que en su ejecución pueda generar datos no realistas o ruido.

Para su implementación, se realizarán los siguientes pasos:
1. Importación del modelo `ADASYN` de la librería **Imbalanced-Learn**.
2. Se muestra la cantidad de datos para cada clase por medio de la función `Counter()` de la librería **Collections**.
3. Se realiza el sobre-muestreo de los datos por medio del modelo `ADASYN` así como de la función `.fit_resample()`.
4. Se muestra la cantidad de datos para cada clase por medio de la función `Counter()` de la librería **Collections**.
5. Se genera un **DataFrame** balanceado con los nuevos datos generados.

````python
#Importación del Modelo
from imblearn.over_sampling import ADASYN

#Impresión de la Cantidad de Datos por Clase Antes del Balanceo
print('Antes:', Counter(y))

#Sobre-Muestreo de los Datos Desbalanceados
adasyn = ADASYN()
X_train, y_train = adasyn.fit_resample(x, y)

#Impresión de la Cantidad de Datos por Clase Después del Balanceo
print('Después:',Counter(y_train))

#Nuevo DataFrame Balanceado
adasyn_df = X_train.copy()
adasyn_df['Clase'] = y_train
````

### Técnica `SMOTE + Tomek Links` para el Balanceo de Datos
La técnica `SMOTE + Tomek Links` combina la habilidad de `SMOTE` para generar datos sintéticos a partir de las clases minoritarias con la habilidad de `Tomek Links` para remover datos de las clases mayoritarias por medio de un método basado en ***Condensed Nearest Neighbors***, esto es, datos de las clases mayoritarias que quedan cercanos a los datos de las clases minoritarias. Una descripción detallada puede ser consultada [en esta liga](https://medium.com/towards-data-science/imbalanced-classification-in-python-smote-tomek-links-method-6e48dfe69bbc).

Para su implementación, se realizarán los siguientes pasos:
1. Importación del modelo `SMOTETomek` de la librería **Imbalanced-Learn**.
2. Se muestra la cantidad de datos para cada clase por medio de la función `Counter()` de la librería **Collections**.
3. Se realiza el sobre-muestreo de los datos por medio del modelo `SMOTETomek` así como de la función `.fit_resample()`.
4. Se muestra la cantidad de datos para cada clase por medio de la función `Counter()` de la librería **Collections**.
5. Se genera un **DataFrame** balanceado con los nuevos datos generados.

````python
#Importación del Modelo
from imblearn.combine import SMOTETomek

#Impresión de la Cantidad de Datos por Clase Antes del Balanceo
print('Antes:', Counter(y))

#Sobre-Muestreo de los Datos Desbalanceados
smote_tom = SMOTETomek(random_state=139)
X_train, y_train = smote_tom.fit_resample(x, y)

#Impresión de la Cantidad de Datos por Clase Después del Balanceo
print('Después:',Counter(y_train))

#Nuevo DataFrame Balanceado
smote_tom_df = X_train.copy()
smote_tom_df['Clase'] = y_train
````

### Técnica `SMOTE + ENN` para el Balanceo de Datos
La técnica `SMOTE + ENN` (***Synthetic Minority Oversampling Technique*** + ***Edited Nearest Neighbor***) es similar a la técnica `SMOTE + Tomek Links`, ya que combina la habilidad de `SMOTE` para generar datos sintéticos a partir de las clases minoritarias con el modelo `ENN` para remover datos de las clases mayoritarias, encuentra el ***K-Nearest Neighbor*** de cada observación y verifica si las clases mayoritarias son las mismas que las clases de la observación, y si son diferentes las elimina. Una descripción detallada puede ser consultada [en esta liga](https://medium.com/towards-data-science/imbalanced-classification-in-python-smote-enn-method-db5db06b8d50).

Para su implementación, se realizarán los siguientes pasos:
1. Importación del modelo `SMOTEENN` de la librería **Imbalanced-Learn**.
2. Se muestra la cantidad de datos para cada clase por medio de la función `Counter()` de la librería **Collections**.
3. Se realiza el sobre-muestreo de los datos por medio del modelo `SMOTEENN` así como de la función `.fit_resample()`.
4. Se muestra la cantidad de datos para cada clase por medio de la función `Counter()` de la librería **Collections**.
5. Se genera un **DataFrame** balanceado con los nuevos datos generados.

````python
#Importación del Modelo
from imblearn.combine import SMOTEENN

#Impresión de la Cantidad de Datos por Clase Antes del Balanceo
print('Antes:', Counter(y))

#Sobre-Muestreo de los Datos Desbalanceados
smote_enn = SMOTEENN()
X_train, y_train = smote_enn.fit_resample(x, y)

#Impresión de la Cantidad de Datos por Clase Después del Balanceo
print('Después:',Counter(y_train))

#Nuevo DataFrame Balanceado
smote_enn_df = X_train.copy()
smote_enn_df['Clase'] = y_train
````

### Gráficos de las Bases de Datos Balanceadas
Finalmente, se muestran los gráficos de los **DataFrames** que fueron ya balanceados por medio del método `.scatterplot()` de la librería **Seaborn**.

````python
#Gráfico de los Resultados
f, axes = plt.subplots(2,2, dpi=200)

sns.scatterplot(x='Serie1', y='Serie2', hue = 'Clase', data=smote_df, ax=axes[0,0])
axes[0,0].set_title('Balanceo con SMOTE')

sns.scatterplot(x='Serie1', y='Serie2', hue = 'Clase', data=adasyn_df, ax=axes[0,1])
axes[0,1].set_title('Balanceo con ADASYN')

sns.scatterplot(x='Serie1', y='Serie2', hue = 'Clase', data=smote_tom_df, ax=axes[1,0])
axes[1,0].set_title('Balanceo con SMOTE + Tomek Links')

sns.scatterplot(x='Serie1', y='Serie2', hue = 'Clase', data=smote_enn_df, ax=axes[1,1])
axes[1,1].set_title('Balanceo con SMOTE + ENN')

plt.setp(axes, yticks=[])
plt.tight_layout()

#Guardar el Gráfico Resultante
#plt.savefig('./Guardados/Balanceados.png', dpi=300, transparent=False, bbox_inches='tight')
plt.show();
````

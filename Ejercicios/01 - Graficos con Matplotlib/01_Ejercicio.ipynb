#### Explicación:
Se realizará el gráfico de diversas **Funciones Trigonométricas** empleando **Matplotlib**. Para ello, primeramente se describe la manera de realizar el gráfico de las funciones ***Seno*** y ***Coseno***:

$$y = sen(\theta)$$
$$y = cos(\theta)$$


Se emplearán las librerías **Matplotlib** y **Math** la cual es un módulo de **Python** que contiene funciones matemáticas. Los pasos a seguir son:

1. Importar las librerías **Math** y **Matplotlib**.
2. Se inicializan dos listas con 360 posiciones (valores del ***0*** al ***359***), todos ellos con valores iguales a ***0***.
3. Se inicializa una lista con valores de los ángulos que van del ***0*** al ***359*** por medio de la función `range()`.
4. Empleando un ciclo `for` se calculan los valores del ***Seno*** y del ***Coseno*** para los ángulos.
   - El ***Seno*** se calcula con el método `math.sin()` donde el valor del ángulo debe estar en radianes ($ángulo*\pi/180$).
   - El ***Coseno*** se calcula con el método `math.cos()` donde el valor del ángulo debe estar en radianes ($ángulo*\pi/180$).
   - El valor ***$\pi$*** se determina con el método `math.pi()`.
5. Se realizan los dos gráficos por medio del método `.plot()` de **Matplotlib**.
6. Se agregan elementos adicionales al gráfico. 

```python
#Importación de Librerías
import math
import matplotlib.pyplot as plt

#Inicialización de Valores
sin = [0] * 360
cos = [0] * 360
ang = [*range(360)]

#Cálculo de las Funciones Trigonométricas
for i in ang:
    sin[i] = math.sin(ang[i]*math.pi/180)
    cos[i] = math.cos(ang[i]*math.pi/180)

#Creación de los Gráficos
plt.plot(ang, sin, color='magenta', marker='', linestyle='-', label='Seno')
plt.plot(ang, cos, color='cyan',    marker='', linestyle='-', label='Coseno')

#Agregación de Título y Etiquetas
plt.legend(loc='best')
plt.title("Funciones Trigonométricas")
plt.xlabel("Ángulo")
plt.ylabel("Valor") 
plt.grid()
plt.show()
```

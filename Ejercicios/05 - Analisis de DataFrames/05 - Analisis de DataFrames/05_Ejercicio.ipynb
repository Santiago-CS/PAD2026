**1.** Cargar las librerías **Pandas**, **Numpy**, **Matplotlib** y **Seaborn**. Posteriormente, crear un **DataFrame** donde se inserte el contenido de la base de datos seleccionada para este ejercicio, mostrando el contenido del **DataFrame** final a través de la función `display()`.

```python
# Código del Reactivo 1

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Lectura de los datos y asignación del nombre de las películas como el índice
airbnb_df = pd.read_csv("./Kaggle/01_AirbnbNY.csv", index_col = 'id')

#Impresión de los Resultados
print("El DataFrame es:")
airbnb_df.head() 
```

**2.** Definir alguna de las columnas como el índice del **DataFrame**.

```python
# Código del Reactivo 2
reviews_col = airbnb_df['number_of_reviews']

#Impresión de los Resultados
print("La columna es del tipo:", type(reviews_col))
print("El tamaño de la columna es:", reviews_col.shape)
reviews_col
```

**3.** Realizar una homogeneización en los encabezados de las columnas empleando el método `.rename()`.

```python
# Código del Reactivo 3
airbnb_df.rename(columns = {
        'neighbourhood': 'Vecindario', 
        'name': 'nombre'
    }, inplace = True)

#Mostrar los Nombres de los Encabezados
airbnb_df.columns
```

**4.** Mostrar la cantidad de celdas que contienen valores nulos en el **DataFrame**.

```python
# Código del Reactivo 4
airbnb_df.isnull()
```

**5.** Debido a que es necesario contar con la información completa para cada columna, es posible realizar la eliminación de los renglones que contienen valores nulos. El siguiente paso es eliminar esos renglones:

```python
# Código del Reactivo 5
no_ren_df = airbnb_df.dropna()

#Verificación de Valores Nulos
no_ren_df.isnull().sum()
```

**6.** Realizar algún proceso de **Imputación** de los ***valores nulos*** restantes en el **DataFrame**:

```python
# Código del Reactivo 6
reviews_mean = airbnb_df['number_of_reviews'].mean()
```

**7.** Determinar el valor de la media de alguna de las columnas e imprimir el resultado con la función `print()`:

```python
# Código del Reactivo 7
#Impresión de los Resultados
print("El valor de la Media de la columna es: %.2f" % reviews_mean)
```

**8.** Realizar un Histograma con los datos de alguna de las columnas contenidas en el **DataFrame**, agregando títulos a los ejes y al gráfico:

```python
# Código del Reactivo 8
plt.figure(figsize=(10, 6))

# Crear el histograma con seaborn
sns.histplot(airbnb_df['number_of_reviews'], bins=40, color='skyblue')

# Agregar los títulos solicitados
plt.title('Distribución del Número de Reseñas', fontsize=14)
plt.xlabel('Número de Reseñas', fontsize=12)
plt.ylabel('Frecuencia (Cantidad de alojamientos)', fontsize=12)

# Mostrar el gráfico
plt.show()
```

**9.** Realizar un gráfico de Correlación del **DataFrame**:

```python
# Código del Reactivo 9

plt.figure(figsize=(10, 8))

# Seleccionar EXCLUSIVAMENTE las columnas numéricas para evitar errores
columnas_numericas = airbnb_df.select_dtypes(include=['float64', 'int64'])

# Calcular la matriz de correlación
matriz_correlacion = columnas_numericas.corr()

# Crear el gráfico de correlación (Heatmap)
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Título del gráfico
plt.title('Gráfico de Correlación de Variables Numéricas', fontsize=14)

# Mostrar el gráfico
plt.show()
```

**10.** Realizar alguna operación de filtrado sobre el **DataFrame**:

```python
# Código del Reactivo 10

# Filtrar alojamientos con más de 100 reseñas
alojamientos_populares = airbnb_df[airbnb_df['number_of_reviews'] > 100]

# Mostrar un mensaje con el resultado
print(f"Se encontraron {alojamientos_populares.shape[0]} alojamientos con más de 100 reseñas.")

# Mostrar las primeras filas del DataFrame ya filtrado
alojamientos_populares.head()
```
